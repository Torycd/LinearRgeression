```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("MILUSA.xlsx")

# Drop rows with missing values in 'ML' or 'GDPR' columns
data = data.dropna(subset=['ML', 'GDPR'])

# Normalize the data
data = (data - data.mean()) / data.std()

Military = data['ML']
Growth = data['GDPR']

def gradient_descent(m_now, b_now, points, L):
    m_gradient = 0
    b_gradient = 0
    n = len(points)
    
    for i in range(n):
        x = points.iloc[i]['ML']
        y = points.iloc[i]['GDPR']
        
        m_gradient += -(2/n) * x * (y - (m_now * x + b_now))
        b_gradient += -(2/n) * (y - (m_now * x + b_now))
    
    m = m_now - L * m_gradient
    b = b_now - L * b_gradient
    
    return m, b

m = 0
b = 0
L = 0.01  # Adjust the learning rate
epochs = 500

for i in range(epochs):
    m, b = gradient_descent(m, b, data, L)
    print(f'Epoch {i + 1}: m = {m}, b = {b}')

plt.scatter(Military, Growth, color="black")
plt.plot(Military, m * Military + b, color="red")
plt.xlabel('Military Expenditure')
plt.ylabel('GDP Growth Rate')
plt.title('Linear Regression using Gradient Descent')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error
import statsmodels.api as sm

data = pd.read_excel("MILUSA.xlsx")

# Drop rows with missing values in 'ML' or 'GDPR' columns
data = data.dropna(subset=['ML', 'GDPR'])

Military = data['ML']
Growth = data['GDPR']

# Add a constant term for the intercept
X = sm.add_constant(Military)  
model = sm.OLS(Growth, X).fit()
predictions = model.predict(X)

print(model.summary())

# Calculate R-squared
r_squared = r2_score(Growth, predictions)
print(f"R-squared: {r_squared}")

# Calculate Mean Squared Error
mse = mean_squared_error(Growth, predictions)
print(f"Mean Squared Error: {mse}")

# Extract coefficients
m = model.params['ML'] 
b = model.params['const']
print(f"Coefficients: m = {m}, b = {b}")

plt.scatter(Military, Growth, color="black")
plt.plot(Military, predictions, color="red")
plt.xlabel('Military Expenditure')
plt.ylabel('GDP Growth Rate')
plt.title('Linear Regression using Statsmodels')
plt.show()
```
